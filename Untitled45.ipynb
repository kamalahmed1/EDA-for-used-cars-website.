# Introduction 
Exploratory Data Analysis (EDA) on data from Hatla2ee, an Egyptian car marketplace for both new and used cars, gives us a deep look into how people buy and sell cars in Egypt. This data has lots of details about cars, like their prices, brands, and features. By studying this data carefully, we can learn about what kinds of cars people like, how much they're willing to pay, and where they're buying them from. This helps us understand the car market in Egypt better, so people can make smarter choices when buying or selling cars.

# Data Cleaning and Preprocessing

```python

df['Mileage'] = df['Mileage'].str.replace(' Km', '').str.replace(',', '').astype(float)

df['Price'] = df['Price'].str.replace(' EGP', '').str.replace(',', '').astype(float)

df['Date Displayed'] = pd.to_datetime(df['Date Displayed'])
```

```python
df.head(2)
```

```python
df.duplicated().sum()
```

```python
# Check the number of rows before and after dropping nulls
rows_before = df.shape[0]
df.dropna(subset=['Price', ], inplace=True)
rows_after = df.shape[0]

print(f"Rows before dropping nulls: {rows_before}")
print(f"Rows after dropping nulls: {rows_after}")
print(f"Number of rows dropped: {rows_before - rows_after}")
```

```python
df.rename(columns={'Mileage': 'KM'}, inplace=True)
```

```python

mean_mileage = df['KM'].mean()

df['KM'] = df['KM'].fillna(mean_mileage)
```

```python
df = df.drop([277, 288])

df = df.drop(df.index[[278, 277]])
```

```python
df.columns
```

# Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

make_counts = df['Make'].value_counts().head(10)

make_counts_df = make_counts.reset_index()
make_counts_df.columns = ['Car Make', 'Frequency']

plt.figure(figsize=(12, 8))

barplot = sns.barplot(data=make_counts_df, x='Frequency', y='Car Make', hue='Car Make', dodge=False, color='blue')
plt.title('Top 10 Most Common Car Makes')
plt.xlabel('Frequency')
plt.ylabel('Car Make')
plt.legend([],[], frameon=False)

for index, value in enumerate(make_counts_df['Frequency']):
    barplot.text(value, index, str(value), color='black', ha="left", va="center")

plt.show()
```

```python
car_counts = df['Name'].value_counts().head(10)  # Get top 10 most common car names

plt.figure(figsize=(12, 8))

car_counts_df = car_counts.reset_index()
car_counts_df.columns = ['Car Name', 'Frequency']


sns.barplot(data=car_counts_df, x='Frequency', y='Car Name', palette='viridis', hue='Car Name', dodge=False)
plt.title('Top 10 Most Common Car Names')
plt.xlabel('Frequency')
plt.ylabel('Car Name')
plt.legend([],[], frameon=False)  
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib.colors as mcolors

colors = ["black", "cyan"]
cmap = mcolors.LinearSegmentedColormap.from_list("Custom", colors)

city_counts = df['City'].value_counts().head(10)
city_counts_df = city_counts.reset_index()
city_counts_df.columns = ['City', 'Frequency']

num_colors = len(city_counts_df)
custom_palette = [cmap(i / num_colors) for i in range(num_colors)]

plt.figure(figsize=(12, 8))
barplot = sns.barplot(data=city_counts_df, x='Frequency', y='City', hue='City', dodge=False, palette=custom_palette, legend=False)
plt.title('Top 10 Most Common Cities')
plt.xlabel('Frequency')
plt.ylabel('City')

for index, value in enumerate(city_counts_df['Frequency']):
    barplot.text(value, index, str(value), color='black', ha="left", va="center")

plt.show()
```

```python

import matplotlib.colors as mcolors

top_20_expensive_cars = df.sort_values(by='Price', ascending=False).head(20)

colors = ["red", "black"]
cmap = mcolors.LinearSegmentedColormap.from_list("Custom", colors)

num_colors = len(top_20_expensive_cars)
custom_colors = [cmap(i / num_colors) for i in range(num_colors)]

plt.figure(figsize=(12, 8))
plt.barh(top_20_expensive_cars['Name'], top_20_expensive_cars['Price'], color=custom_colors)
plt.xlabel('Price')
plt.ylabel('Car Name')
plt.title('Top 20 Most Expensive Cars')
plt.gca().invert_yaxis()
plt.show()

```

```python
import matplotlib.pyplot as plt

labels = ['With', 'Without']
colors = ['black', 'cyan']

plt.figure(figsize=(10, 6))
plt.subplot(1, 3, 1)
air_conditioner_counts = df['Air Conditioner'].value_counts()
plt.pie(air_conditioner_counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Air Conditioner Availability')

plt.subplot(1, 3, 2)
power_steering_counts = df['Power Steering'].value_counts()
plt.pie(power_steering_counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Power Steering Availability')

plt.subplot(1, 3, 3)
remote_control_counts = df['Remote Control'].value_counts()
plt.pie(remote_control_counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Remote Control Availability')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

city_stats = df.groupby('City').agg({'Price': 'mean', 'Make': 'count'}).reset_index()
city_stats = city_stats.rename(columns={'Make': 'Car Frequency'})

city_stats = city_stats.sort_values(by='Price', ascending=False)

plt.figure(figsize=(60, 30))
sns.barplot(x='Price', y='City', data=city_stats, color='skyblue')
plt.title('Average Price of Used Cars Across Cities')
plt.xlabel('Average Price')
plt.ylabel('City')
plt.show()

```

**average price of used cars across cities" refers to the mean price at which used cars are being sold in different cities. It provides insights into the general pricing trends of used cars in various geographic locations**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a violin plot to visualize the distribution of prices for each car make
plt.figure(figsize=(43, 25))
sns.violinplot(x='Make', y='Price', data=df)
plt.title('Price Distribution by Car Make')
plt.xlabel('Car Make')
plt.ylabel('Price')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt

X = df[['KM', 'Age']]
y = df['Price']

rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
rf_regressor.fit(X, y)

feature_importances = rf_regressor.feature_importances_

feature_names = X.columns

sorted_indices = feature_importances.argsort()[::-1]
sorted_feature_importances = feature_importances[sorted_indices]
sorted_feature_names = feature_names[sorted_indices]

plt.figure(figsize=(10, 6))
plt.bar(sorted_feature_names, sorted_feature_importances)
plt.title('Feature Importances')
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.xticks(rotation=45)
plt.show()
```
